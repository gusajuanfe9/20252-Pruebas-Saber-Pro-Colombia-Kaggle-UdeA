### <font color='46B8A9'> **2.2. Distribuci√≥n y participaci√≥n de variables categ√≥ricas**

```python
'''Para cada variable categ√≥rica se crea un DataFrame con los valores √∫nicos de la columna como √≠ndice, donde la primera
   columna corresponde a la frecuencia absoluta (n√∫mero de veces que se repite cada valor o categor√≠a), y la segunda columna
   muestra la frecuencia relativa (porcentaje que representa cada categor√≠a respecto al total de observaciones en esa columna)'''

columnas1 = df.select_dtypes(include=['object'])  # Selecciona las columnas categ√≥ricas (tipo 'object') del DataFrame

# Crear un nuevo DataFrame con frecuencia absoluta y frecuencia relativa de cada categor√≠a
for columna in columnas1:
    df_count = pd.DataFrame({ "Conteo": df[columna].value_counts(),
                             "Proporci√≥n": df[columna].value_counts(normalize=True).mul(100).round(2).astype(str) + ' %'})
    print(df_count)  # Imprime el DataFrame con las frecuencias
    print('-'*70)
```

### <font color='46B8A9'> **2.6. An√°lisis relevantes**

```python
import plotly.express as px

# Contar los valores √∫nicos en la columna 'E_HORASSEMANATRABAJA' y convertir el resultado en un DataFrame
df_count = df['E_HORASSEMANATRABAJA'].value_counts().reset_index()

# Renombrar las columnas del DataFrame resultante
df_count.columns = ['E_HORASSEMANATRABAJA', 'count']

# Ordenar el DataFrame de mayor a menor seg√∫n la cantidad de estudiantes
df_count = df_count.sort_values('count', ascending=False)

# Crear gr√°fico de barras
fig = px.bar(
    df_count,
    x='E_HORASSEMANATRABAJA',
    y='count',
    text='count',
    labels={
        'E_HORASSEMANATRABAJA': 'Horas que trabaja a la semana',
        'count': 'Cantidad de estudiantes'
    },
    title='Distribuci√≥n de estudiantes seg√∫n las horas que trabajan a la semana'
)

# Configurar dise√±o del gr√°fico
fig.update_traces(textposition='outside')
fig.update_layout(
    xaxis_title='Horas que trabaja a la semana',
    yaxis_title='N√∫mero de estudiantes',
    title_x=0.5,
    coloraxis_showscale=False
)

# Mostrar el gr√°fico
fig.show()
```

La mayor√≠a de los estudiantes trabaja m√°s de 30 horas a la semana, lo que evidencia que una parte considerable de la poblaci√≥n estudiantil combina sus estudios con una jornada laboral extensa. En comparaci√≥n, los grupos que trabajan menos horas representan proporciones m√°s reducidas, siendo especialmente bajos los que dedican menos de 10 horas semanales. Esta distribuci√≥n sugiere que para muchos estudiantes el trabajo ocupa un papel importante en su vida cotidiana, lo que podr√≠a influir en la disponibilidad de tiempo para sus actividades acad√©micas y, en algunos casos, en su rendimiento acad√©mico.
